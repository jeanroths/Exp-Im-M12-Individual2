```python
from bark import generate_audio, preload_models
import numpy as np
import IPython.display as ipd

# Pré-carregue os modelos
preload_models()

# Texto a ser convertido em fala
text = "Bem-vindo(a) ao Summoner's Rift, novato(a)! Pode relaxar, eu vou te mostrar tudo o que você precisa saber. Vamos nessa!"

# Gere o áudio
audio_array = generate_audio(text)

# Exiba o áudio diretamente no notebook
ipd.Audio(audio_array, rate=24000)
```

```python
text = "Welcome to Summoner's Rift new player! Relax, I'll show you everything you must know to become a legend. Let's go!"
voice_preset = "v2/en_speaker_9"  # Substitua pelo estilo desejado
audio_array = generate_audio(text, history_prompt=voice_preset)

# Reproduza o áudio
ipd.Audio(audio_array, rate=24000)
```

```python
from scipy.io.wavfile import write

# Salve o áudio em um arquivo WAV
write("fala_vi.wav", 24000, (audio_array * 32767).astype(np.int16))

# Baixe o arquivo para o seu computador
from google.colab import files
files.download("fala_vi.wav")
```
